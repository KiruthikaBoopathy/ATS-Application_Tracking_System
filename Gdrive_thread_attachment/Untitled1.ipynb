```python
import imaplib as imp
import email
from datetime import datetime
import html2text
import mysql.connector

def fetch_email_content(imapUserEmail, imapPassword):
    imapHostServer = 'imap.gmail.com'
    imapVar = imp.IMAP4_SSL(imapHostServer)

    imapVar.login(imapUserEmail, imapPassword)
    imapVar.select('new_client')
    tmp, data = imapVar.search(None, 'UNSEEN')

    email_texts = []

    for n in data[0].split():
        typ, msg_data = imapVar.fetch(n, '(RFC822)')
        msg = email.message_from_bytes(msg_data[0][1])

        html_content = None

        for part in msg.walk():
            if part.get_content_type() == "text/html":
                html_content = part.get_payload(decode=True).decode(part.get_content_charset(), 'ignore')

        if html_content:
            plain_text_content = html2text.html2text(html_content)
        else:
            plain_text_content = "No HTML content found in the email."

        email_texts.append(plain_text_content)

    return email_texts

def process_and_insert(email_texts, db_config):
    conn = mysql.connector.connect(**db_config)
    cursor = conn.cursor()

    for email_text in email_texts:
        Ref_no = title = start_date = end_date = bill_rate = of_openings = position_type = location = description = None
        lines = email_text.split('\n')

        for line in lines:
            parts = line.split(":", 1)
            if len(parts) == 2:
                keyword, value = parts[0].strip(), parts[1].strip()

                if keyword == "Ref No":
                    Ref_no = value
                elif keyword == "Title":
                    title = value
                elif keyword == "Start Date":
                    try:
                        date_obj = datetime.strptime(value, '%m/%d/%Y')
                        start_date = date_obj.strftime('%d/%m/%y')
                    except ValueError:
                        print(f"Error: Invalid date format - {value}")
                elif keyword == "End Date":
                    try:
                        date_obj = datetime.strptime(value, '%m/%d/%Y')
                        end_date = date_obj.strftime('%d/%m/%y')
                    except ValueError:
                        print(f"Error: Invalid date format - {value}")
                elif keyword == "Bill Rate":
                    bill_rate = value  #.replace('$', ' ')
                elif keyword == "# of Openings":
                    of_openings = value
                elif keyword == "Position Type":
                    position_type = value
                elif keyword == "Locations":
                    location = value
                elif keyword == "Description":
#                     value = email_text.split("Description")[1]
                    description = value

        email_data = (Ref_no, title, start_date, end_date, bill_rate, of_openings, position_type, location, description)
        insert_email(cursor, email_data)

        conn.commit()
        conn.close()


def insert_email(cursor, email_data):
    insert_query = "INSERT INTO suppliers (Ref_no, title, start_date, end_date, bill_rate, of_openings, position_type, location, description) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)"
    cursor.execute(insert_query, email_data)

if __name__ == '__main__':
    imapUserEmail = 'kiruthika.b@vrdella.com'
    imapPassword = 'renm kixf nlxy avbx'

    db_config = {
        "host": "localhost",
        "user": "root",
        "password": "vrdella!6",
        "database": "email_extraction",
    }

    email_texts = fetch_email_content(imapUserEmail, imapPassword)
    process_and_insert(email_texts, db_config)
print("Email extraction and insertion completed successfully.")






```
