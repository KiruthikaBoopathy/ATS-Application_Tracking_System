```python
import requests
import concurrent.futures

# Define the URLs of the sample APIs you want to test

def fetch_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    return {'error': f'Failed to retrieve data from {url}'}

if __name__ == '__main__':
    api_urls = ['https://jsonplaceholder.typicode.com/posts/1',
                'https://jsonplaceholder.typicode.com/posts/2',
                'https://jsonplaceholder.typicode.com/posts/3',
                'https://jsonplaceholder.typicode.com/posts/4']

    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(fetch_data, api_urls))

    for url, result in zip(api_urls, results):
        print(f"Data from {url}: {result}")
```

```python
import asyncio
import requests
import nest_asyncio

nest_asyncio.apply()  # Enable running asyncio within Jupyter Notebook

async def fetch_data(url):
    response = await asyncio.to_thread(requests.get, url)
    if response.status_code == 200:
        return response.json()
    return {'error': f'Failed to retrieve data from {url}'}

async def main():
    api_urls = ['https://jsonplaceholder.typicode.com/posts/1',
                'https://jsonplaceholder.typicode.com/posts/2',
                'https://jsonplaceholder.typicode.com/posts/3',
                'https://jsonplaceholder.typicode.com/posts/4']

    tasks = [fetch_data(url) for url in api_urls]
    results = await asyncio.gather(*tasks)

    for url, result in zip(api_urls, results):
        print({"data": result})

if __name__ == '__main__':
    asyncio.run(main())
```

```python
import requests
import multiprocessing

# Define a function to fetch data from a URL
def fetch_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    return {'error': f'Failed to retrieve data from {url}'}

if __name__ == '__main__':
    api_urls = [
        'https://jsonplaceholder.typicode.com/posts/1',
        'https://jsonplaceholder.typicode.com/posts/2',
        'https://jsonplaceholder.typicode.com/posts/3',
        'https://jsonplaceholder.typicode.com/posts/4'
    ]

    with multiprocessing.Pool(processes=len(api_urls)) as pool:
        results = pool.map(fetch_data, api_urls)

    for url, result in zip(api_urls, results):
        print({"data": result})
```

```python
import httpx

async def fetch_data(url):
    async with httpx.AsyncClient() as client:
        response = await client.get(url)
        if response.status_code == 200:
            return response.json()
        return {'error': f'Failed to retrieve data from {url}'}

if __name__ == '__main__':
    import asyncio

    api_urls = [
        'https://jsonplaceholder.typicode.com/posts/1',
        'https://jsonplaceholder.typicode.com/posts/2',
        'https://jsonplaceholder.typicode.com/posts/3',
        'https://jsonplaceholder.typicode.com/posts/4'
    ]

    async def main():
        tasks = [fetch_data(url) for url in api_urls]
        results = await asyncio.gather(*tasks)

        for url, result in zip(api_urls, results):
            print({"data": result})

    asyncio.run(main())
```

```python
import aiohttp
import asyncio

async def fetch_data(url):
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as response:
            if response.status == 200:
                return await response.json()
            return {'error': f'Failed to retrieve data from {url}'}

if __name__ == '__main__':
    api_urls = [
        'https://jsonplaceholder.typicode.com/posts/1',
        'https://jsonplaceholder.typicode.com/posts/2',
        'https://jsonplaceholder.typicode.com/posts/3',
        'https://jsonplaceholder.typicode.com/posts/4'
    ]

    async def main():
        tasks = [fetch_data(url) for url in api_urls]
        results = await asyncio.gather(*tasks)

        for url, result in zip(api_urls, results):
            print({"data": result})

    asyncio.run(main())
```
